```
#Square of a number: Create a function that returns the square of a number.
def square(a):
  a = a**2
  return(a)

print(square(2))
print(square(3))
```

```
#Cube of a number: Create a function that returns the cube of a number.
def cube(a):
  a = a**3
  return(a)

print(cube(2))
print(cube(3))
```

```
#Absolute value: Create a function that returns the absolute value of a number.
def absolute(a):
  if a > 0:
    return(a)
  if a < 0:
    return(-a)

print(absolute(2))
print(absolute(-4))
```

```
#Factorial of a number: Create a function that returns the factorial of a number.
def factorielle(n):
   if n == 0 :
      return 1
   else:
      Fact = 1
      for i in range(2,n+1):
         Fact += (Fact * i)
      return Fact

print(factorielle(-2))
print(factorielle(6))
```

```
#Mode of a list of numbers: Create a function that returns the mode of a list of numbers, 
#for instance [68, 99, 65, 44, 77, 44, 44] --> 44.

def modeListe(list):
  dico = {}
  for k,v in enumerate(list):
    if v not in dico.keys():
      dico[v] = 1
    else:
      dico[v] += 1

  maxk = 0
  maxv = 0
  for k,v in dico.items():
    if v > maxv:
      maxv = v
      maxk = k
  print(dico)
  return(f"Le mode est : {maxk}, la récurrence est de : {maxv}")

print(modeListe([68, 99, 65, 44, 77, 44, 44]))
print(modeListe([2, 2, 1, 3, 3, 3, 4]))


#Ou avec fonction1 :


# test = [1,1,2,3,4,5,3,4,5,5,5]

# dico = {k : test.count(k) for k in test}

# print(dico)
# max_freq = max({v for v in dico.values()})
# print(max_freq)


#Ou avec fonction2 :


# def modeListeNombre(a):
#   return(max(set(a), key = a.count))

# print(modeListeNombre([68, 99, 65, 44, 77, 44, 44]))
# print(modeListeNombre([68, 99, 65, "text", "text", "text", 44]))




```

```
#Average of a list of numbers: Create a function that returns the average of a list of numbers.
def moyenneListeNombre(a):
  somme = 0
  diviseur = 0
  for k,v in enumerate(a):
    somme += v
    diviseur += 1
  return(somme/diviseur)

a = [68, 99, 65, 44, 77, 44, 44]
print(moyenneListeNombre(a))
a = [12, 11, 55, 53, 23, 11]
print(moyenneListeNombre(a))
  
```

```
#Minimum of a list of numbers: Create a function that returns the minimum from within a list of numbers.
def mini(a):
  minimum = a[0]
  for k,v in enumerate(a):
    if v < minimum:
      minimum = v
  return(minimum)

a = [68, 99, 65, 44, 30, 44, 44]
print(mini(a))
a = [5, 10, 10, 10, 5]
print(mini(a))
```

```
#Maximum of a list of numbers: Create a function that returns the maximum from within a list of numbers.
def maxi(a):
  maximum = a[0]
  for k,v in enumerate(a):
    if v > maximum:
      maximum = v
  return(maximum)

a = [68, 99, 65, 44, 30, 44, 44]
print(maxi(a))
a = [5, 10, 10, 10, 5]
print(maxi(a))
```
